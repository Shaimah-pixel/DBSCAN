```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

# Filling missing values

```python
data.culmen_length_mm=data.culmen_length_mm.fillna(data.culmen_length_mm.median())
```

```python
data.culmen_depth_mm=data.culmen_depth_mm.fillna(data.culmen_depth_mm.median())
```

```python
data.flipper_length_mm=data.flipper_length_mm.fillna(data.flipper_length_mm.median())
```

```python
data.body_mass_g=data.body_mass_g.fillna(data.body_mass_g.median())
```

```python
data.sex=data.sex.fillna("sex")
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

# Converting categorical column to numerical

```python
data.sex=le.fit_transform(data.sex)
```

```python
data.dtypes
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
df_data=ss.fit_transform(data)
df_data
```

```python
db=DBSCAN(eps=0.5,min_samples=3) #dbscan has 2 hyper paratmeter, distance which is epsilon and minimum datapoints which is min sample
#to remove noise,increase the distance and decrease the data points
```

```python
model_db=db.fit(df_data)
model_db
```

```python
cc=set(model_db.labels_)  #how many clusters or groups are formed..-1 indicates noise
cc
```

```python
print(cc)
```

```python
len(cc)
```

```python
list_labels=list(model_db.labels_) #to determine how many datapoints are in noise
list_labels.count(-1)
```

```python
c1=pd.Series(list_labels)
c1
```

```python
df_data=pd.concat([data,c1],axis=1) #to check datapoints assigned to cluster
df_data
```

```python
df_data.rename(columns={df_data.columns[5]:"cluster"},inplace=True)
df_data
```
